```python
import graph_tool.all as gt
import graph_parser
import utils
import numpy as np
np.set_printoptions(threshold=np.nan)

from IPython.html.widgets import interactive
from IPython.display import display
from os import listdir
from os.path import join, isfile
```

```python
def browse_dimacs():
    instances_path = "instances"
    files = [ join(instances_path,f) for f in listdir(instances_path) if isfile(join(instances_path,f)) ]
    def view_graph(f):
        global g
        g = graph_parser.parse_col(f)
        pos = gt.sfdp_layout(g)
        gt.graph_draw(g, pos=pos, output_size=(500,500), inline=True, vertex_text=g.vertex_index)
        return g
    w = interactive(view_graph, f=files)
    display(w)
    return w.result
```
